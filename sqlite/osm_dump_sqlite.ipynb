```python
from pathlib import Path
import osmiter
from itertools import islice
```

```python
PBF_PATH = Path("~/data/osm_de_oct2023.pbf").expanduser()
DB_PATH = Path("~/data/osm_de_oct2023.db").expanduser()
```

```python
!sqlite-utils create-database $DB_PATH --enable-wal --init-spatialite
```

```python
# open DB_PATH as IO[bytes]
pbf_buf = PBF_PATH.open("rb")
```

```python
f = [feature for feature in islice(osmiter.iter_from_pbf_buffer(pbf_buf), 200)]
```
